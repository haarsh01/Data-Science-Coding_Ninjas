### matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, \*\* kwargs)

plot([x], y, [fmt], *, data=None, **kwargs)

plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

***The coordinates of the points or line nodes are given by x, y.***

The optional parameter **fmt** is a convenient way for defining basic formatting like color, marker and linestyle. It's a shortcut string notation described in the Notes section below.

`plot(x, y)        # plot x and y using default line style and color
 plot(x, y, 'bo')  # plot x and y using blue circle markers
 plot(y)           # plot y using x as index array 0..N-1
 plot(y, 'r+')     # ditto, but with red plusses`

You can use Line2D properties as keyword arguments for more control on the appearance. Line properties and fmt can be mixed. The following two calls yield identical results:

`plot(x, y, 'go--', linewidth=2, markersize=12)
 plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)`
 
 
**fmt** : *str, optional*

A format string, e.g. 'ro' for red circles. See the Notes section for a full description of the format strings.

Format strings are just an abbreviation for quickly setting basic line properties. All of these and more can also be controlled by keyword arguments.

This argument cannot be passed as keyword.

 [Read more on it here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

### Bubble Charts

In scatter plot one can pass size of bubbles to signify another quantity.

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

```python
plt.scatter(year, salary)
plt.show()
```

```python
plt.scatter(year, salary, s = 100) # Bubble size to be 100
plt.show()
```

```python
plt.scatter(year, salary, s = population) # Bubble size to signify population
plt.title("Bubble Chart")
plt.show()
```

```python
# Alpha used to set opaqness of bubbles
plt.scatter(year, salary, s = population, alpha = 0.4) # Bubble size to signify population
plt.title("Bubble Chart")
plt.show()
```

```python
col = np.arange(len(year))

plt.scatter(year, salary, s = population, alpha = 0.4, c = col) # Bubble size to signify population
plt.title("Bubble Chart")
plt.show()
```

```python
# Edge or color of bubble boundary can also be changed

col = np.arange(len(year))

plt.scatter(year, salary, s = population, alpha = 0.4, c = col, edgecolors='r') # Bubble size to signify population
plt.title("Bubble Chart")
plt.show()
```

```python
# Markers can also be changed, say to triangles

col = np.arange(len(year))

plt.scatter(year, salary, s = population, alpha = 0.4, c = col, marker = '^', edgecolors='r') # Bubble size to signify population
plt.title("Bubble Chart")
plt.show()
```

```python
# Writing text is not straightforward. You can use plt.text(x, y) function

col = np.arange(len(year))

plt.scatter(year, salary, s = population, alpha = 0.4, c = col, marker = '^', edgecolors='r') # Bubble size to signify population
plt.title("Bubble Chart")
plt.text(year[0] + 0.1, salary[0] + 0.1, population[0])
plt.show()
```

```python
# Writing text is not straightforward. You can use plt.text(x, y) function

col = np.arange(len(year))

plt.scatter(year, salary, s = population, alpha = 0.4, c = col, edgecolors='r') # Bubble size to signify population
plt.title("Bubble Chart")

for i in range (len(year)):
    plt.text(year[i] + 0.1, salary[i] + 0.1, population[i])
plt.show()
```

### Pie Graphs

```python
sizes = [3, 4, 6, 2]

plt.pie(sizes)
plt.show()
```

```python
sizes = [3, 4, 6, 2]

plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
# Changing colors

sizes = [3, 4, 6, 2]
color = ["blue", "purple", "red", "pink"]

plt.pie(sizes, colors = color)

plt.axis("equal")
plt.show()
```

```python
# Adding Labels

labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
color = ["blue", "purple", "red", "pink"]
plt.title("Split among classes")
plt.pie(sizes, colors = color, labels = labels)

plt.axis("equal")
plt.show()
```

```python
# Print percentage split

labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
color = ["blue", "purple", "red", "pink"]
plt.title("Split among classes")
plt.pie(sizes, colors = color, labels = labels, autopct = "%.2f%%") # Can pass function or just pass a string

plt.axis("equal")
plt.show()
```

```python
# Highlight a particular class using explode property

labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
color = ["blue", "purple", "red", "pink"]
explode = [0.2, 0.1, 0, 0] # By default clockwise

plt.title("Split among classes")
plt.pie(sizes, colors = color, explode = explode, labels = labels, autopct = "%.2f%%") # Can pass function or just pass a string
# counterclock = False to make it go clockwise
plt.axis("equal")
plt.show()
```

```python
# Changing start angle

labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
color = ["blue", "purple", "red", "pink"]
explode = [0.1, 0, 0, 0]

plt.title("Split among classes")
plt.pie(sizes, counterclock = False, startangle = 100, colors = color, explode = explode, labels = labels, autopct = "%.2f%%") # Can pass function or just pass a string
plt.axis("equal")
plt.show()
```

### Histogram

```python
b = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 6, 7, 8, 8, 20, 22, 11, 14]
```

```python
plt.hist(b)
plt.show()
```

```python
plt.hist(b, edgecolor = "black")
plt.show()
```

```python
# xticks

xt = np.arange(23)

plt.hist(b, edgecolor = "black")
plt.xticks(xt)
plt.show()
```

```python
# By default 10 bins are made
```

```python
# change number of bins


# width of each bin is (total bins) / (number of bins)

# bins follow [a, b) convetion (a included, b not included)

xt = np.arange(23)

plt.hist(b, bins = 25, edgecolor = "black")
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b, bins = [1, 3, 6, 10, 15, 20], edgecolor = "black")
plt.show()
```

```python
# Axis ranges can be changed

xt = np.arange(23)

plt.hist(b, bins = [3, 5, 7, 10, 15], edgecolor = "black")
plt.axis([0, 25, 0, 10])
plt.show()
```
