```python
import sqlite3
import pandas as pd

db = sqlite3.connect('School.sqlite')
```

```python
# pd.read_sql_query(query, connectionObject)

data = pd.read_sql_query('Select * from Student', db)

data
```

```python
# tuples with age = 30

data[data['Age'] == 21]
```

```python
# Creating new tables with Pandas

data = pd.DataFrame([
         [201, "Aman", "English", 20000], 
         [202, "Komal", "Math", 22000],
         [203, "Mahesh", "Science", 23000]
        ], columns = ["ID", "Name", "Course", "Salary"])

data
```

```python
# Converting DataFrame into a Table

data.to_sql("Faculty", db)
```

```python
# Add Column in table

data = pd.read_sql_query('Select * from Student', db)
data["Address"] = "Test"

data
```

```python
data.to_sql("Student", db, if_exists = "replace")
```
